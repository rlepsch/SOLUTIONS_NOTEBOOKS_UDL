```python
import numpy as np
import matplotlib.pyplot as plt
```

First we start with a base probability density function

```python
# Define the base pdf
def gauss_pdf(z, mu, sigma):
  pr_z = np.exp( -0.5 * (z-mu) * (z-mu) / (sigma * sigma))/(np.sqrt(2*3.1413) * sigma)
  return pr_z
```

```python
z = np.arange(-3,3,0.01)
pr_z = gauss_pdf(z, 0, 1)

fig,ax = plt.subplots()
ax.plot(z, pr_z)
ax.set_xlim([-3,3])
ax.set_xlabel('$z$')
ax.set_ylabel('$Pr(z)$')
plt.show();
```

Now let's define a nonlinear function that maps from the latent space $z$ to the observed data $x$.

```python
# Define a function that maps from the base pdf over z to the observed space x
def f(z):
    x1 = 6/(1+np.exp(-(z-0.25)*1.5))-3
    x2 = z
    p = z * z/9
    x = (1-p) * x1 + p * x2
    return x

# Compute gradient of that function using finite differences
def df_dz(z):
    return (f(z+0.0001)-f(z-0.0001))/0.0002
```

```python
x = f(z)
fig, ax = plt.subplots()
ax.plot(z,x)
ax.set_xlim(-3,3)
ax.set_ylim(-3,3)
ax.set_xlabel('Latent variable, $z$')
ax.set_ylabel('Observed variable, $x$')
plt.show()
```

Now let's evaluate the density in the observed space using equation 16.1




The first term is the inverse of the determinant of the D  D Jacobian matrix f[z, ]/z, which contains terms fi[z, ]/zj at position (i, j).

```python
# TODO -- plot the density in the observed space
# Replace these line
x = np.arange(-3,3,0.01) 
pr_x = (1 / abs(df_dz(z))) * pr_z
```

```python
# Plot the density in the observed space
fig,ax = plt.subplots()
ax.plot(x, pr_x)
ax.set_xlim([-3,3])
ax.set_ylim([0, 0.5])
ax.set_xlabel('$x$')
ax.set_ylabel('$Pr(x)$')
plt.show();
```

Now let's draw some samples from the new distribution (see section 16.1) - we draw $z^{*}$ from the base density and pass this through the function so that $x^{*}$ = f($z^{*}$)

```python
np.random.seed(1)
n_sample = 20

# TODO -- Draw samples from the modeled density
# Replace this line
x_samples = np.random.choice(z, size=n_sample, replace=False)
```

```python
# Draw the samples
fig,ax = plt.subplots()
ax.plot(x, pr_x)
for x_sample in x_samples:
  ax.plot([x_sample, x_sample], [0,0.1], 'r-')

ax.set_xlim([-3,3])
ax.set_ylim([0, 0.5])
ax.set_xlabel('$x$')
ax.set_ylabel('$Pr(x)$')
plt.show();
```
